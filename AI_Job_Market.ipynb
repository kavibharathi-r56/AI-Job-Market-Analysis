# Stage 3 – EDA and Visualizations

## UNIVARIANT ANALYSIS

### Chart 1 – Salary Distribution

```python
plt.figure()
sns.histplot(df['salary_usd'], kde=True)
plt.title("Salary Distribution")
plt.show()
```

Interpretation:
- The chart shows how salaries are distributed across AI jobs.
- This analysis uses the salary_usd feature to understand how compensation varies across the dataset
- Most salaries are concentrated in a mid-range, with a few very high salaries indicating a right-skewed distribution.

## Chart 2- Job Role Count

```python
plt.figure()
sns.countplot(y='job_title', data=df, order=df['job_title'].value_counts().index[:10])
plt.title("Top 10 Job Roles")
plt.show()
```

Interpretation:
- Shows most frequent job roles
- Feature used: job_title
- Insight: Roles like Data Scientist and ML Engineer dominate hiring

## Chart 3 – Experience Level Distribution

```python
plt.figure()
sns.countplot(x='experience_level', data=df)
plt.title("Experience Level Distribution")
plt.show()
```

Interpretation:
- Shows distribution of experience levels
- Feature: experience_level
- Insight: Majority of jobs are mid to senior level

## BIVARIANT ANALYSIS

## Chart 4 – Salary vs Experience


```python
plt.figure()
sns.boxplot(x='experience_level', y='salary_usd', data=df)
plt.title("Salary vs Experience Level")
plt.show()
```

Interpretation:
- Shows salary variation by experience
- Features: experience_level, salary_in_usd
- Insight: Salary increases significantly with experience

## Chart 5 – Salary by Company Size

```python
plt.figure()
sns.barplot(x='company_size', y='salary_usd', data=df)
plt.title("Salary by Company Size")
plt.show()
```

Interpretation:
- Shows relationship between company size and salary
- Features: company_size, salary_in_usd
- Insight: Large companies tend to offer higher salaries

## Chart 6 – Correlation Heatmap

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Interpretation:
- Shows relationships between numeric variables
- Feature: all numeric columns
- Insight: Helps identify which variables influence salary

## MULTIVARIATE ANALYSIS

## Chart 7 – Salary by Experience & Employment Type

```python
plt.figure()
sns.barplot(x='experience_level', y='salary_usd', hue='employment_type', data=df)
plt.title("Salary by Experience & Employment Type")
plt.show()
```

Interpretation:
- Combines 3 variables
- Features: experience_level, employment_type, salary_usd
- Insight: Full-time roles consistently earn higher salaries

## Chart 8 – Pairplot

```python
sns.pairplot(df.select_dtypes(include=['number']))
plt.show()
```

Interpretation:
- Shows relationships among multiple numeric variables
- Insight: Helps detect patterns and outliers

## Chart 9 – Pivot Table

```python
pivot = df.pivot_table(values='salary_usd',
                       index='job_title',
                       columns='experience_level',
                       aggfunc='mean')

pivot.head()
```

Interpretation:
- Shows average salary across roles and experience
- Insight: Some roles show faster salary growth than others
